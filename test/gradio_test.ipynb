```python
from collections import defaultdict

# è™•ç† ingredients ç‚ºæ¢åˆ—æ ¼å¼å­—ä¸²
def format_ingredients(ingredients):
    grouped = defaultdict(list)
    for item in ingredients:
        group_name = item['group'] or ""
        grouped[group_name].append(item)

    result = []
    for group_name, items in grouped.items():
        result.append(f"{group_name}å…§çš„ææ–™:")
        for idx, item in enumerate(items, 1):
            result.append(f"{idx}. {item['name']} : {item['amount']}")
        result.append("")  # ç©ºè¡Œåˆ†éš”ç¾¤çµ„
    return "\n".join(result)
```

```python
def get_recipe_block(recipes):
    # å‹•æ…‹ç”¢å‡º HTML æ”¾é€² gallery å®¹å™¨
    html = '<div class="image-gallery">'
    for idx, r in enumerate(recipes):
        html += f"""
        <div class="image-option" data-index="{idx}">
            <img src="{r['image_url']}" alt="{r['recipe_name']}">
            <span>{r['recipe_name']}</span>
        </div>
        """
    html += '</div>'

    return html
```

```python
def generate(recipe, question):
    system_prompt = """
    ä½ æ˜¯ä¸€ä½æ“…é•·åˆ†æé£Ÿè­œçš„å°ˆæ¥­å»šå¸«ï¼Œå°å„é¡æ–™ç†çš„ææ–™ã€è£½ä½œæ­¥é©Ÿèˆ‡çƒ¹é£ªæŠ€å·§éå¸¸ç†Ÿæ‚‰ã€‚
    è«‹æ ¹æ“šæˆ‘æä¾›çš„æ¢ä»¶èˆ‡è³‡è¨Šï¼Œçµ¦å‡ºæ˜ç¢ºä¸”è©³ç´°çš„é£Ÿè­œåˆ†æçµæœã€‚
    ç„¡è«–æˆ‘æå‡ºä½•ç¨®è¦æ±‚æˆ–æè¿°ä½•ç¨®æƒ…å¢ƒï¼Œä½ éƒ½éœ€ä¾æ“šæ¢ä»¶æä¾›å°ˆæ¥­çš„å»ºè­°èˆ‡è§£èªªã€‚
    å›ç­”æ™‚è«‹ä½¿ç”¨ç¹é«”ä¸­æ–‡ã€‚
    """

    # å¥—ç”¨é€² user_prompt
    formatted_ingredients = format_ingredients(recipe['ingredients'])

    user_prompt = f"""
    ä»¥ä¸‹æ˜¯æˆ‘è¦ä½ åˆ†æçš„é£Ÿè­œå…§å®¹:
    é£Ÿè­œåç¨±: {recipe['recipe_name']}
    é£Ÿè­œé€£çµ: {recipe['recipe_link']}
    é£Ÿè­œä½œè€…: {recipe['user_name']}
    é£Ÿè­œä½œè€…é£Ÿè­œæ•¸é‡: {recipe['user_recipe']}
    é£Ÿè­œä½œè€…ç²‰çµ²æ•¸é‡: {recipe['user_fans']}
    é£Ÿè­œåœ–ç‰‡: {recipe['image_url']}
    é£Ÿè­œæè¿°: {recipe['description']}
    é£Ÿè­œæ¨™ç±¤: {recipe['tags']}
    é£Ÿè­œå–œæ­¡æ•¸: {recipe['like_count']}
    é£Ÿè­œä¸€èµ·åšæ•¸: {recipe['together_count']}
    é£Ÿè­œç•™è¨€æ•¸: {recipe['comment_count']}
    é£Ÿè­œé£Ÿæ:
    {formatted_ingredients}
    é£Ÿè­œæ­¥é©Ÿ: {recipe['steps']}
    é£Ÿè­œè§€çœ‹æ•¸: {recipe['view_count']}
    é£Ÿè­œæ—¥æœŸ: {recipe['date']}

    è«‹æ ¹æ“šé€™äº›è³‡è¨Šï¼Œå›ç­”å•é¡Œ:
    {question}

    å›ç­”æ™‚è«‹ä½¿ç”¨ç¹é«”ä¸­æ–‡ã€‚
    """


    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=1,
        max_tokens=300
    )

    thoughts = response.choices[0].message.content
    return thoughts
```

```python
def handle_chat_question(selected_indexes_str, question):
    if not selected_indexes_str.strip():
        return "âš ï¸ è«‹å…ˆé¸æ“‡è‡³å°‘ä¸€å€‹é£Ÿè­œã€‚"

    try:
        selected_indexes = [int(i) for i in selected_indexes_str.split(",") if i.strip().isdigit()]
    except:
        return "âš ï¸ è§£æé¸å–è³‡æ–™å¤±æ•—"

    selected_recipes = [recipes[i] for i in selected_indexes if 0 <= i < len(recipes)]

    if not selected_recipes:
        return "âš ï¸ æ‰¾ä¸åˆ°é¸å–çš„é£Ÿè­œ"

    output_html = ['<div class="chat-box">']
    for r in selected_recipes:
        answer = generate_result(r, question)
        output_html.append(f"""
        <div class="chat-item">
            <img class="chat-thumb" src="{r['image_url']}" alt="{r['recipe_name']}">
            <div class="chat-info">
                <h3 class="chat-title">{r['recipe_name']}</h3>
                <div class="chat-desc">{answer}</div>
            </div>
        </div>
        """)
    output_html.append("</div>")

    return "\n".join(output_html)
```

```python
def generate_true_false(recipe, question):
    system_prompt = """
    ä½ æ˜¯ä¸€ä½å°ˆæ¥­ä¸”ç¶“é©—è±å¯Œçš„å»šå¸«ï¼Œæ“…é•·åˆ†æå„é¡é£Ÿè­œçš„ææ–™ã€è£½ä½œæ­¥é©Ÿèˆ‡çƒ¹é£ªæŠ€å·§ã€‚  
    è«‹æ ¹æ“šæˆ‘æä¾›çš„æ¢ä»¶èˆ‡è³‡è¨Šï¼Œåˆ¤æ–·ä¸¦å›ç­”ã€Œtrueã€æˆ–ã€Œfalseã€ï¼Œä»¥æ˜ç¢ºå›æ‡‰è©²å•é¡Œã€‚  
    å›ç­”æ™‚è«‹åƒ…è¼¸å‡ºã€Œtrueã€æˆ–ã€Œfalseã€ã€‚ã€‚è«‹å‹¿è¼¸å‡ºå…¶ä»–æ–‡å­—ã€‚ä¸è¦åŠ ä¸Šä»»ä½•è§£é‡‹ã€‚
    """

    # å¥—ç”¨é€² user_prompt
    formatted_ingredients = format_ingredients(recipe['ingredients'])

    user_prompt = f"""
    ä»¥ä¸‹æ˜¯æˆ‘è¦ä½ åˆ†æçš„é£Ÿè­œå…§å®¹:
    é£Ÿè­œåç¨±: {recipe['recipe_name']}
    é£Ÿè­œé€£çµ: {recipe['recipe_link']}
    é£Ÿè­œä½œè€…: {recipe['user_name']}
    é£Ÿè­œä½œè€…é£Ÿè­œæ•¸é‡: {recipe['user_recipe']}
    é£Ÿè­œä½œè€…ç²‰çµ²æ•¸é‡: {recipe['user_fans']}
    é£Ÿè­œåœ–ç‰‡: {recipe['image_url']}
    é£Ÿè­œæè¿°: {recipe['description']}
    é£Ÿè­œæ¨™ç±¤: {recipe['tags']}
    é£Ÿè­œå–œæ­¡æ•¸: {recipe['like_count']}
    é£Ÿè­œä¸€èµ·åšæ•¸: {recipe['together_count']}
    é£Ÿè­œç•™è¨€æ•¸: {recipe['comment_count']}
    é£Ÿè­œé£Ÿæ:
    {formatted_ingredients}
    é£Ÿè­œæ­¥é©Ÿ: {recipe['steps']}
    é£Ÿè­œè§€çœ‹æ•¸: {recipe['view_count']}
    é£Ÿè­œæ—¥æœŸ: {recipe['date']}

    è«‹æ ¹æ“šé€™äº›è³‡è¨Šï¼Œå›ç­”å•é¡Œ:
    {question}

    è«‹åªå›è¦†ã€Œtrueã€æˆ–ã€Œfalseã€ï¼Œå…¶ä»–ä»»ä½•æ–‡å­—éƒ½æ˜¯éŒ¯èª¤çš„ã€‚
    """


    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0,
        max_tokens=5
    )

    # æ¨¡å‹è¼¸å‡º
    raw_output = response.choices[0].message.content.strip().lower()

    # ä½¿ç”¨æ­£å‰‡å¼å°‹æ‰¾ "true" æˆ– "false"
    match = re.search(r'\b(true|false)\b', raw_output)

    if match:
        return match.group(1)
    else:
        return "false"
```

```python
def handle_true_false_question(selected_indexes_str, question):
    if not selected_indexes_str.strip():
        return "âš ï¸ è«‹å…ˆé¸æ“‡è‡³å°‘ä¸€å€‹é£Ÿè­œã€‚"

    try:
        selected_indexes = [int(i) for i in selected_indexes_str.split(",") if i.strip().isdigit()]
    except:
        return "âš ï¸ è§£æé¸å–è³‡æ–™å¤±æ•—"

    selected_recipes = [recipes[i] for i in selected_indexes if 0 <= i < len(recipes)]

    if not selected_recipes:
        return "âš ï¸ æ‰¾ä¸åˆ°é¸å–çš„é£Ÿè­œ"

    true_items = []
    false_items = []

    for r in selected_recipes:
        answer = generate_true_false(r, question)
        block = f'''
        <div class="match-item">
            <img class="match-thumb" src="{r['image_url']}" alt="{r['recipe_name']}">
            <div class="match-info">
                <h3 class="match-title">{r['recipe_name']}</h3>
            </div>
        </div>
        '''
        # <div class="match-desc">{answer}</div> å¯åŠ å…§æ–‡

        if answer == "true":
            true_items.append(block)
        else:
            false_items.append(block)

    output_html = [f'''<div class="match-group">''']

    if true_items:
        output_html.append(f'''
        <div class="match-section match-true">
            <h2 class="match-header">âœ… ç¬¦åˆæ¢ä»¶çš„é£Ÿè­œ</h2>
            <div class="true-false-image">
                <div class="items-container">
                    {''.join(true_items)}
                </div>
            </div>
        </div>
        ''')

    if false_items:
        output_html.append(f'''
        <div class="match-section match-false">
            <h2 class="match-header">âŒ ä¸ç¬¦åˆæ¢ä»¶çš„é£Ÿè­œ</h2>
            <div class="true-false-image">
                <div class="items-container">
                    {''.join(false_items)}
                </div>
            </div>
        </div>
        ''')

    output_html.append(f'''</div>''')

    return "\n".join(output_html)
```
